```python
from rescue_ai_model import (
    DisasterRescueModel, EnvironmentSnapshot, VictimDetection, InjuryLevel, EntryPoint
)
model = DisasterRescueModel()
environment = EnvironmentSnapshot(0.6, 0.5, 0.45, 0.3)
detections = model.process_stream([
    VictimDetection('V-1', 20, 80, 0.3, 0.2, 0.85, 0.70, InjuryLevel.CRITICAL),
    VictimDetection('V-2', 60, 65, 0.7, 0.4, 0.4, 0.65, InjuryLevel.SEVERE),
])
priorities = model.prioritize_victims(detections, environment)
points = [
    EntryPoint('North Gate', 10, 95, 0.25, 0.30, 0.75, 0.70, 20.0),
    EntryPoint('Service Tunnel', 5, 40, 0.45, 0.20, 0.80, 0.65, 35.0),
]
best = model.suggest_entry_point(points, priorities, environment)
priorities, best
```
